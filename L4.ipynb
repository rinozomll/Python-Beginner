```python
import numpy as np
import matplotlib.pyplot as plt
import numdifftools as nd

# 定義函數
def f(x):
    return 3 * x**2 + 1

# 數值導數函數
d = nd.Derivative(f)

# 測試單點導數
print(d(2))  # 應該接近 12

# 產生數值區間
y = np.linspace(-3, 3, 100)
fy = f(y)                 # f(x) 在每個點的值
dy = np.array([d(val) for val in y])  # 對每個點計算導數

# 繪圖
plt.plot(y, fy, label='f(x) = 3x² + 1')
plt.plot(y, dy, label="f'(x)", linestyle='--')
plt.legend()
plt.title('Function and its Derivative')
plt.grid(True)
plt.show()
```
